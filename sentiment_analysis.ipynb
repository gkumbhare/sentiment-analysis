```python
avg_sentence_length = []
Percentage_of_Complex_words = []
Fog_Index = []
complex_word_count = []
avg_syllable_word_count = []

def measure(file_content):
    # Remove punctuations
    file_content = re.sub(r'[^\w\s.]', '', file_content)

    # Split the given text file into sentences
    sentences = file_content.split('.')

    # Total number of sentences in a file
    num_sentences = len(sentences)

    # Total words in the file
    words = [word for word in file_content.split() if word.lower() not in stopwords_set]
    num_words = len(words)

    syllable_count = 0
    syllable_words = []
    complex_words = []

    for word in words:
        # Handle exceptions for words ending with "es" or "ed"
        if word.endswith('es') or word.endswith('ed'):
            word = word[:-2]

        vowels = 'aeiou'
        syllable_count_word = sum(1 for letter in word if letter.lower() in vowels)

        if syllable_count_word >= 1:
            syllable_words.append(word)
            syllable_count += syllable_count_word

        # Identify complex words
        if syllable_count_word > 2:
            complex_words.append(word)

    avg_sentence_len = num_words / num_sentences
    avg_syllable_word_count = syllable_count / len(syllable_words)
    Percent_Complex_words = len(complex_words) / num_words
    Fog_Index = 0.4 * (avg_sentence_len + Percent_Complex_words)

    return avg_sentence_len, Percent_Complex_words, Fog_Index, len(complex_words), avg_syllable_word_count

# Iterate through each file or doc
for file_name in os.listdir(text_dir):
    with open(os.path.join(text_dir, file_name), 'r') as file:
        content = file.read()
        x, y, z, a, b = measure(content)
        avg_sentence_length.append(x)
        Percentage_of_Complex_words.append(y)
        Fog_Index.append(z)
        complex_word_count.append(a)
        avg_syllable_word_count.append(b)
```

```python
word_count = []
average_word_length = []

def cleaned_words(file_content):
    # Remove punctuations
    file_content = re.sub(r'[^\w\s]', '', file_content)

    # Extract cleaned words by removing stop words
    words = [word for word in file_content.split() if word.lower() not in stopwords_set]

    # Calculate word count and average word length
    length = sum(len(word) for word in words)
    avg_length = length / len(words) if len(words) > 0 else 0

    return len(words), avg_length

# Iterate through each file
for file_name in os.listdir(text_dir):
    with open(os.path.join(text_dir, file_name), 'r') as file:
        content = file.read()
        x, y = cleaned_words(content)
        word_count.append(x)
        average_word_length.append(y)
```

```python
def count_personal_pronouns(file_content):
    personal_pronouns = ["I", "we", "my", "ours", "us"]
    count = 0
    for pronoun in personal_pronouns:
        count += len(re.findall(r'\b' + pronoun + r'\b', file_content, re.IGNORECASE))  # \b is used to match word boundaries
    return count

pp_count = []

for file_name in os.listdir(text_dir):
    with open(os.path.join(text_dir, file_name), 'r') as file:
        content = file.read()
        # Check for 'US' as a country name and remove it from the text content
        content = re.sub(r'\bUS\b', '', content)
        count = count_personal_pronouns(content)
        pp_count.append(count)
```
